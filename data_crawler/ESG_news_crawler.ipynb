```python
import pandas as pd
import numpy as np
import os
import datetime as datetime
```

```python
df_news_actv = pd.read_csv('/content/drive/MyDrive/넘블 나이스 디앤비 챌린지/2. ESG raw data/news_actv_pre.csv')
df_news_clos = pd.read_csv('/content/drive/MyDrive/넘블 나이스 디앤비 챌린지/2. ESG raw data/news_clos_pre.csv')

# 불필요열 삭제
df_news_actv = df_news_actv.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1)
df_news_clos = df_news_clos.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1)
```

```python
ESG_news_actv = pd.read_csv('/content/drive/MyDrive/넘블 나이스 디앤비 챌린지/2. ESG raw data/ESG_news_actv.csv')
ESG_news_clos = pd.read_csv('/content/drive/MyDrive/넘블 나이스 디앤비 챌린지/2. ESG raw data/ESG_news_clos.csv')
```

```python
# ESG_news_clos와 날짜 인덱싱하기 전의 news_clos_pre와 결합 후 >> 날짜 기준으로 인덱싱

news_clos_esg = pd.concat([df_news_clos,ESG_news_clos],axis=1)

news_clos_esg = news_clos_esg[news_clos_esg['pubDate'].str.contains('2018|2019|2020|2021')]
news_clos_esg = news_clos_esg[~news_clos_esg['pubDate'].str.contains('▲ 낙찰업체')]
news_clos_esg
```

```python
# 날짜 기준 인덱싱
df_news_actv = df_news_actv[df_news_actv['pubDate'].str.contains('2018|2019|2020|2021')]
df_news_actv = df_news_actv[~df_news_actv['pubDate'].str.contains('▲ 낙찰업체')]
```

```python
df_news_actv
```

```python
ESG_news_actv = ESG_news_actv.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1)
```

```python
df_news_actv = df_news_actv.reset_index()
```

```python
df_news_actv
```

```python
news_actv_esg = pd.concat([df_news_actv,ESG_news_actv],axis=1)
news_actv_esg
```

```python
news_actv_esg.to_csv('/content/drive/MyDrive/넘블 나이스 디앤비 챌린지/2. ESG raw data/news_actv_esg.csv')
news_clos_esg.to_csv('/content/drive/MyDrive/넘블 나이스 디앤비 챌린지/2. ESG raw data/news_clos_esg.csv')
```
